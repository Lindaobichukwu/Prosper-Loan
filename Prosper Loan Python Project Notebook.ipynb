```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns




df = pd.read_csv('Prosper Loan.csv')

plt.scatter(data = df, x = 'LoanOriginalAmount', y = 'IncomeRange', color = 'brown')
plt.title('Loan Original Amount By Income Range')
plt.xlabel('Loan Original Amount')
plt.ylabel('Income Range');
```

```python
df.LoanOriginalAmount.describe()
```

>>The chart above is a Bivariate Exploration which shows insights to Loan Original Amount 
>>and the Income range column using a Scatter plot. 

>>I wanted to know if Income affected the amount of Loan been granted to borrowers.

>>From the plot, the people with the Income range of $50,000 - 74,999, $75,000 - 99,999 and $100,000+ had more Loan amount given to them. 
>>The people with the least Loan Amount assigned to them where the borrowers who earn $0, next is the Unemployed.

>>I observed from the scatter plot above, that Income truely affects the amount of Loan individuals are been granted in Prosper Loan Company because the more your Income, the more Loan you will be granted. The less your income, the less loan you will be awarded.

```python
plt.scatter(data = df, x = 'Investors', y = 'LoanStatus', color = 'tomato')
plt.title('Number of Investors By Loan Status')
plt.xlabel('Number of Investors')
plt.ylabel('Loan Status');
```

>>The chart above is a Bivariate Exploration which shows insights to Number of Investors and the Loan Status column using a 
Scatter plot.

>> I wanted to know if Number of Investors affected the current status of the Loan.

>>Fron the Scatter plot above, it is been shown that the Loan status that were Cancelled, had the lowest amount of investors. Followed by the Past Due (>120 days). Third Lowest is the Final Payment In Progress etc.
>>The Loan Status of the Completed had the most number of Investors. Second highest were the Defaulted and then the Charged Off.

>>The Visualization and Insights from the plot above shows that, the more investors you get to fund your Loan, the better chances your Loan Status will be Completed but if you have less investors to fund your Loan, your Loan Status will be Cancelled.

```python
sns.regplot(data=df, x='Term', y='LoanOriginalAmount', color = 'indigo').set(title = 'Loan Original Amount By Term', xlabel = 'Term', ylabel = 'Loan Original Amount');
```

```python
df.Term.describe()
```

```python
df.Term.value_counts()
```

>>The chart above is a Bivariate Exploration which shows insights to Loan Original Amount and  the Term(The length of the loan expressed in months) column using a Reg plot.

>>I wanted to know if Loan original amount had any relationship with the lenght of the loan expressed in months.

>>The Reg plot above shows that there are only three Terms to consider(12 months, 36 months and 60 months). 
>>Less Loan was assigned to the 12 months Term borrowers, followed by the 60 months borrowers. The Term 36 months had the most amount of Loan assigned to people.

>>The insight from the plot above is that the more your Loan Amount the more Term you have to pay back and the Less your Loan amount the less months you have to pay back the Loan owed.
